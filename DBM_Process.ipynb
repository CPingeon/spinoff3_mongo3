---

#### *Cration d'un "MongoClient", instance pour se connecter  la BDD par "mongod", et assignation des paramtres (BDD en local ("localhost") et port par dfaut ("27017")*

```python
from pymongo import MongoClient

client = MongoClient()
```

```python
client = MongoClient('localhost', 27017)
```

---

#### *On choisit le nom de la BDD ("spinoff") et de la collection ("movies")*

```python
db = client['spinoff']
```

```python
collection = db['movies']
```

---

#### *Cration de la dfinition et de la variable indispensables pour insrer les donnes*

```python
def lire_CSV_Dict(nom_fichier):
    data = []
    with open(nom_fichier, newline='', encoding='utf8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
            
    return data
```

```python
movies = lire_CSV_Dict("data/movies.csv")
```

---

#### *Insertion des donnes*

```python
collection.insert_many(movies)
```

---

#### *Vrification de la cration de la BDD puis de la collection*

```python
client.list_database_names()
```

```python
db.list_collection_names()
```

---

#### *Requte permettant l'affichage des 10 premiers documents*

```python
for movie in db.movies.find().limit(10):
    print(movie)
```
